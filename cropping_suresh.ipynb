```python

%matplotlib inline
#The line above is necesary to show Matplotlib's plots inside a Jupyter Notebook

import cv2
from matplotlib import pyplot as plt
import pandas as pd
import glob
```

```python
raghul_csv = pd.read_csv("./Dataset/key_points_suresh.csv",header = 0)
raghul_csv = raghul_csv[["Frames#","leftWrist_x","leftWrist_y","rightWrist_x","rightWrist_y","leftWrist_score","rightWrist_score"]]
raghul_csv.head
```

```python
len(raghul_csv)
```

```python
list_of_files = glob.glob("./Dataset/Alphabets/Suresh/*.jpg")
```

```python
for i in range(len(raghul_csv.values)):
    img = cv2.imread(list_of_files[i]) # reads image 'opencv-logo.png' as grayscale
# plt.imshow(img)
    (x,y) = (int(raghul_csv.iloc[i]["leftWrist_x"]),int(raghul_csv.iloc[i]["leftWrist_y"]))
# cv2.rectangle(img, (points[0]-50, points[1]-50), (points[0]+50, points[1]+50), (255,0,0))
    img = cv2.rectangle(img, (x-356,y-484),(x+228,y+100), (255,0,0))
#     plt.imshow(img)
#     plt.show()
    (x,y) = (int(raghul_csv.iloc[i]["rightWrist_x"]),int(raghul_csv.iloc[i]["rightWrist_y"]))
# cv2.rectangle(img, (points[0]-50, points[1]-50), (points[0]+50, points[1]+50), (255,0,0))
#     img = cv2.rectangle(img, (x-356,y-484),(x+228,y+100), (0,255,0))
    plt.imshow(img)
    plt.show()
#     cv2.imshow("test", img)
#     cv2.waitKey(0)
#     cv2.destroyAllWindows()
    print("RED: (LEFT) : ",raghul_csv.iloc[i]["leftWrist_score"],"GREEN: (RIGHT) : ",raghul_csv.iloc[i]["rightWrist_score"])
```
